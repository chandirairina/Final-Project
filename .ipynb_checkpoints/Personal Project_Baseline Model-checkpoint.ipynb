```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("dataset_baseline.csv")
```

```python
df[df["Travel Destination"]=="Tuban"]
```

```python
df=df.drop_duplicates(["Travel Destination","Location","Review Count"],keep="first", inplace=False)
```

```python
df=df.reset_index().drop("index",axis=1)
```

```python
cv = CountVectorizer()
cv_result = cv.fit_transform(df['Things to Do'])
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
cv_df = pd.DataFrame(cv_result.todense(),columns= cv.get_feature_names(),index=df['Travel Destination'])
cv_df
```

```python
cos_sin = cosine_similarity(cv_result)
```

```python
pd.DataFrame(cos_sin, index=df["Travel Destination"], columns=df["Travel Destination"])
```

```python
df[df['Travel Destination'] == "Ubud"].index[0]
```

```python
pd.Series(cos_sin[3])
```

```python
cv_df.info()
```

```python
cv_df
```

```python
def get_recomendation(place):
    index_to_search = df[df['Travel Destination'] == place].index[0]
    place_similar = pd.Series(cos_sin[index_to_search])
    index_similar = place_similar.sort_values(ascending=False).index
    
#     for i in 
#     return df.loc[index_similar]['Travel Destination']
    return index_similar
```

```python
df[df['Travel Destination'] == "Seoul"].index[0]
```

```python
cos_sin[df[df['Travel Destination'] == "Seoul"].index[0]]
```

```python
df[(df["Country Name"]=="Indonesia") & (df["Travel Destination"]=="Jakarta")]
```

```python
get_recomendation("Seoul")
```

```python
prefer=get_recomendation('Seoul')
```

```python
recomm_id=[]

for i in prefer:
    if (df["Country Name"].iloc[i])=="Indonesia":
        recomm_id.append (i)
```

```python
for i in recomm_id[0:10]:
    print(df.loc[i]["Travel Destination"],"---",df.loc[i]["Location"])
```
