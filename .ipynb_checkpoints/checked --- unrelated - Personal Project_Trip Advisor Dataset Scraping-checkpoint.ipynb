Perlu untuk mengubah Review Count menjadi data numerik. Agar kemudian data-data yang duplikat dapat ditambahkan Reviewnya satu sama lain.

```python
indoreview["reviewcount"][0][0:-17]
```

```python
def numerik(x):
    return x["reviewcount"][0:-17]
```

```python
indoreview["reviewcount"]=indoreview.apply(numerik, axis=1)
```

```python
indoreview["reviewcount"]=indoreview["reviewcount"].str.replace(".","")
```
