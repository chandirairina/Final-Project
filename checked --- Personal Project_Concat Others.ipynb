there are some wrong data (USA location, while we actually only scrap for asia)

```python
cek=[]

for i in range(len(df_others)):
    if " Amerika Serikat" in df_others["Location"][i].split(","):
        cek.append("Yes")
    else:
        cek.append("No")

len(cek)
```

```python
df_others["Wrong Data_USA"]=cek
```

```python
df_others=df_others[df_others["Wrong Data_USA"]=="No"]
```

```python
df_others=df_others.reset_index().drop(["index","Wrong Data_USA"], axis=1)
```

```python
df_others.to_csv("dataset_others.csv", index=False)
```
