## Recsys1 : pre-processing

### Menyesuaikan simbol-simbol agar tidak mengganggu dalam proses cleaning data

```python
df1["Things to Do"]=df1["Things to Do"].str.replace("[","").str.replace("]","").str.replace("'","").str.replace(" ","").str.replace("(",",(").str.replace(",Lihatsemua,",",").str.replace(",Lihatsemua","").str.replace("&","dan").str.replace("-","")
```

```python
df1
```

```python
df1["Things to Do"]=df1["Things to Do"].str.split(",")
```

```python
df1
```

```python
todo=[]

for j in range(len(df1)):
    inside_todo=[]
    
    for i in df1["Things to Do"][j]:
        if "(" in i:
            inside_todo.append(" ")
        else:
            inside_todo.append(i)
    todo.append(inside_todo)
```

```python
len(todo)
```

```python
df1["Things to Do"]=todo
```

```python
df1
```

```python
for i in range(len(df1)):
    df1["Things to Do"][i]=''.join([str(elem) for elem in df1["Things to Do"][i]])
```

```python
df1
```

```python
df1=df1.rename(columns={"location":"Location",
                  "reviewcount":"Review Count"})
```

```python
df1.to_csv("dataset_recsys1_engineered.csv", index=False)
```

```python
df1=pd.read_csv("dataset_recsys1_engineered.csv")
```

```python
cv = CountVectorizer()
cv_todo_result = cv.fit_transform(df1['Things to Do'])
```

```python
cv.get_feature_names()
```

```python
todrop=[
 'cilandak',
 'gambir',
 'grogolpetamburan',
 'jatinegara',
 'kebayoranbaru',
 'mampangprapatan',
 'menteng',
 'pademangan',
 'palmerah',
 'pasarminggu',
 'penjaringan',
 'sawahbesar',
 'setiabudi',
 'tanahabang',
 'tebet']
```

```python
df1
```
